```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips=sns.load_dataset('tips')
tips
```

```python
sns.get_dataset_names()
```

```python
sns.histplot(tips["total_bill"])
```

```python
sns.distplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'],kde=False)
```

```python
sns.distplot(tips['total_bill'],bins=30)
```

```python
sns.scatterplot(x=tips.total_bill,y=tips['tip'],data=tips,style='smoker')
#hue is a parameter that can differentiate the different types of data you have
```

```python
sns.scatterplot(x=tips.total_bill,y=tips.tip,data=tips,hue='size')
```

```python
sns.scatterplot(x=tips.total_bill,y=tips['tip'],data=tips,style='smoker',hue='time')
```

```python
#categorical plot 
sns.catplot(x='day',y='tip',data=tips)
```

```python
sns.catplot(x='sex',y='total_bill',data=tips)
```

```python
sns.barplot(x='day',y='total_bill',data=tips)
```

```python
#kind of scatter plot but it will give distribution
sns.jointplot(x='total_bill',y='tip',data=tips)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
sns.jointplot(x = tips.total_bill, y=tips.tip, data = tips, kind="hex")
```

```python
sns.jointplot(x = tips.total_bill, y=tips.tip, data = tips, kind="kde") #contour>>meterological
```

```python
#pairplot>> it will plot all the pairwise relationship of the dataframe
sns.pairplot(tips)
```

```python
sns.pairplot(tips,hue='sex')
```

```python
sns.pairplot(tips, hue = 'smoker',palette = 'coolwarm')
```

```python
#Boxplot >> to know outliers
#1, 2, 3, 4, 5, 100 >>> extreme data, for each day give me outlier for total bill
sns.boxplot(x = 'day', y='total_bill', data=tips, palette = 'rainbow')
```
