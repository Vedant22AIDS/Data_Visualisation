```python
#Que1
plt.figure()
x=range(1,50)
y=range(1,50)
plt.title("Draw a line")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.ylim(0,160)
plt.xlim(0,50)
plt.plot(x,y)
plt.show()
```

```python
#Que3 
plt.figure(figsize=(6,4))
x1=[10,20,30]
y1=[20,40,10]
x2=[10,20,30]
y2=[40,10,30]
plt.title("Two or more lines on same plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.plot(x1,y1,color='blue',label='line 1')
plt.plot(x2,y2,color='green',label='line 2')
plt.ylim(10,40,10)
plt.xlim(10,30,10)
plt.legend()
plt.show()
```

```python
#Que4 
plt.figure(figsize=(6,4))
x1=[10,20,30]
y1=[20,40,10]
x2=[10,20,30]
y2=[40,10,30]
plt.title("Two or more lines on same plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.plot(x1,y1,color='blue',linewidth=3,label='line 1-width-3')
plt.plot(x2,y2,color='green',linewidth=5,label='line 2width-5')
plt.ylim(10,40,10)
plt.xlim(10,30,10)
plt.legend()
plt.show()
```

```python
#Que6
# data to plot
n_groups = 5
men_means = (22, 30, 33, 30, 26)
women_means = (25, 32, 30, 35, 29)

# create plot
fig, ax = plt.subplots()
index = np.arange(n_groups)
bar_width = 0.35
opacity = 0.8

rects1 = plt.bar(index, men_means, bar_width,
alpha=opacity,
color='g',
label='Men')

rects2 = plt.bar(index + bar_width, women_means, bar_width,
alpha=opacity,
color='r',
label='Women')

plt.xlabel('Person')
plt.ylabel('Scores')
plt.title('Scores by person')
plt.xticks(index + bar_width, ('G1', 'G2', 'G3', 'G4', 'G5'))
plt.legend()

plt.tight_layout()
plt.show()
```
