```python
plt.plot(x, np.sin(x-0), color = 'blue')
plt.plot(x, np.sin(x-1), color = 'g')
plt.plot(x, np.sin(x-2), color = (1.0, 0.5, 0)) #rgb
```

```python
#changing the limits of x and y axis
plt.plot(x, np.sin(x-0), color = 'blue')
plt.ylim(-1.5, 1.5)
plt.xlim(1, 15)
plt.show()
```

```python
#removing empty spaces from plot from all sides
plt.plot(x, np.sin(x-0), color = 'blue')
plt.ylim(-1.5, 1.5)
plt.xlim(1, 15)
plt.axis('equal')
plt.show()
```

```python
plt.plot(x, np.sin(x-0), color = 'blue')
plt.ylim(-1.5, 1.5)
plt.xlim(1, 15)
plt.axis('equal')
plt.grid()
plt.show()
```

```python
#Adding title to the plot
plt.plot(x, np.sin(x-0), color = 'blue')
plt.ylim(-1.5, 1.5)
plt.xlim(1, 15)
plt.title("This is sin plot before real data execution to understand the syntax")
plt.axis('equal')
plt.show()
```

```python
#Giving label to x and y axis
plt.plot(x, np.sin(x-0), color = 'blue')
plt.ylim(-1.5, 1.5)
plt.xlim(1, 15)
plt.title("This is sin plot before real data execution to understand the syntax")
plt.axis('equal')
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

```python
plt.plot(x, np.sin(x-0), color = 'blue',label='SinX')
plt.plot(x, np.cos(x-0), color = 'red',label='CosX')
plt.ylim(-1.5, 1.5)
plt.xlim(1, 15)
plt.title("This is sin plot before real data execution to understand the syntax")
#plt.axis('equal')
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend() #A legend is an area describing the elements of the graph.inside the plot
plt.show()
```

```python
#categorical data
x=['a','b','c','d']
y=[3,6,5,4]
```

```python
x
```

```python
y
```

```python
plt.figure(figsize=(6,3))
plt.bar(x,y,label='x',color='orange')
plt.title("Categorical Data")
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.ylim(1,10)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(6,3))
plt.barh(x,y,label='x',color='yellow')#Horizontally
plt.title("Categorical Data")
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
plt.show()
```
