```python
#Merge and join
df1=pd.DataFrame({'Key1':[1,2,4,5,6],'Key2':[4,5,6,7,0],'Key3':[3,4,5,6,6]})
df1
```

```python
df2=pd.DataFrame({'Key1':[1,2,45,6,67],'Key4':[56,5,6,7,8],'Key5':[3,56,5,6,6]})
df2
```

```python
#left join - all the elements of df1 and common elements of df2
pd.merge(df1,df2,how='left')
```

```python
#right join - all the elements of df2 and common elements of df1
pd.merge(df1,df2,how='right')
```

```python
#outer join - all the keys of both dataframe
pd.merge(df1,df2,how='outer')
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1,df2,how='cross')# all possible cominations(cartesian product) all df1 rows with df2 rows
```

```python
pd.merge(df1,df2,how='left',left_on='Key2',right_on='Key4')
#common elements between key2 and key4 and perform left join
#both dataframe cannot have same column name hence it is key1_x and key1_y
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1,df2,how='right',left_on='Key2',right_on='Key4')
```
